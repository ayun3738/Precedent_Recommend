```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/miri/의료관련판결문.csv')
df
```

```python
df['sen_list'] = df['전처리'].apply(lambda x : x.split('.'))
df
```

```python
sentences = []
num_sen = []

for i, sen_list in enumerate(df['sen_list'].tolist()):
  sentences.extend(sen_list)
  num_sen.extend([i] * (len(sen_list)))
  # print(len(sen_list))
  # print([i] * (len(sen_list)))

len(sentences), len(num_sen)
```

```python
df_dataset = pd.DataFrame(zip(sentences, num_sen), columns= [['sentence', 'pan_num']])  
df_dataset
```

```python
for i in range(10):
  print(df_dataset.iloc[i]['sentence'])
  pan_num = df_dataset.iloc[i]['pan_num']
  print(f'해당 판례 번호: {pan_num}')
```
