```python
import pandas as pd
import re
from sentence_transformers import SentenceTransformer
```

```python
NUM_SAMPLES = 200

df = pd.read_pickle('/content/drive/MyDrive/NLP/mingyu/가공데이터/의료사고결과.pickle')
df = df[:NUM_SAMPLES]
df
```

```python
def preprocess(sen):

  #한글과 구두점을 제외하고 '가,나,다,라,마,바,사,아' 문자와 특수문자, 숫자 제거
  sen = re.sub(r'\n', r' ', sen)
  sen = re.sub('사건개요', '', sen)
  sen = re.sub('기초사실', '', sen)
  sen = re.sub(r'[0-9]+\.', r'', sen)
  sen = re.sub(r'\([^)]*\)', r'', sen)
  sen = re.sub(r' [가나다라마바사아자차]\s?\.', r'', sen)
  sen = re.sub(r'[^ㄱ-ㅎ가-힣\.]+', r' ', sen)
  sen = re.sub(r' [가나다라마바사아자차] ', r' ', sen)

  #단어와 구두점, 쉼표 사이에 공백
  sen = re.sub(r'([\.])', r' \1', sen)

  #다수 개 공백을 하나의 공백으로 치환
  sen = re.sub(r"\s+", " ", sen)
  s_split = sen.split('.')

  return_s = [s for s in s_split if len(s)>3]

  return return_s
```

# inference

```python
similarity_list = df['similarity_list'].tolist()
```

```python
test_sen = df.iloc[198]['sen_list']
test_emb = model.encode(test_sen[0])
test_emb_src = model_src.encode(test_sen[0])
```

```python
from numpy.linalg import norm
import numpy as np
```

```python
def cal_cosine(sen, t_list):
  cosine = lambda A, B : np.dot(A,B)/(norm(A)*norm(B))
  similarities = [cosine(sen, target) for target in t_list]

  sorted_c = sorted(similarities, reverse= True)
  if len(similarities) > 5:
    return sorted_c[:5]
  else:
    return sorted_c
```

```python
cosine_list = []
cosine_list_src = []

for similarity in similarity_list:
  cosine_list.append(cal_cosine(test_emb, similarity))
  cosine_list_src.append(cal_cosine(test_emb_src, similarity))

df['top5'] = cosine_list
df['top5_s'] = cosine_list_src
```

```python
df['mean'] = df['top5'].apply(lambda x : sum(x)/len(x))
df['max'] = df['top5'].apply(lambda x : max(x))
df['rank'] = df['mean'].rank() + df['max'].rank()

idx = np.argmax(df['rank'])
pan = df['이유'][idx]
```

```python
df['mean_src'] = df['top5_s'].apply(lambda x : sum(x)/len(x))
df['max_src'] = df['top5_s'].apply(lambda x : max(x))
df['rank_src'] = df['mean_src'].rank() + df['max_src'].rank()

idx_src = np.argmax(df['rank_src'])
pan_src = df['이유'][idx_src]
```

```python
simi = df.iloc[idx]['top5']

display(f'의뢰사연: {test_sen[0]}')
print(f'유사한 판결문 번호: {idx}')
print(f'유사한 판결문 유사성: {simi}')
print(f'유사한 판결문 원본: {pan}')
```
